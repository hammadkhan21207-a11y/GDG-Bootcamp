```python
df["Item"].fillna(df["Item"].mode()[0], inplace=True)
df["Quantity"].fillna(df["Quantity"].mean, inplace=True)
df["Price Per Unit"].fillna(df["Price Per Unit"].mode()[0], inplace=True)
df["Total Spent"].fillna(df["Total Spent"].mode()[0], inplace=True)
df["Payment Method"].fillna(df["Payment Method"].mode()[0], inplace=True)
df["Location"].fillna(df["Location"].mode()[0], inplace=True)
df["Transaction Date"].fillna(df["Transaction Date"].mode()[0], inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df["Total Spent"] = pd.to_numeric(df["Total Spent"], errors="coerce")
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Quantity", y="Total Spent")
sns.regplot(data=df, x="Quantity", y="Total Spent", scatter=True, color="red")
plt.title("Relationship between Quantity and Total Spent")
plt.xlabel("Quantity")
plt.ylabel("Total Spent ($)")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include="number")
corr_matrix = numeric_cols.corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation between numerical columns")
plt.show()

print(corr_matrix)
```

```python
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"], errors="coerce")

monthly_revenue = df.groupby(df["Transaction Date"].dt.to_period("M"))["Total Spent"].sum()

monthly_revenue.plot(kind="bar", figsize=(10,5), color="skyblue")
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Total Revenue (dollars)")
plt.xticks(rotation=45)
plt.show()
```

for a cafe month is a meaningful time unit beacuse a lot of cafes revenue after a month.This helps the cafe's owner which month is more profitable and which is less profitable
