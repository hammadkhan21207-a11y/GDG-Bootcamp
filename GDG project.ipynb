STEP 5 $ Model Building + Evaluation

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor  # <-- Changed

# Columns
onehot_cols = ['Gender', 'Internet_Access', 'City', 'Department']
ordinal_cols = ['Education_Level', 'Attendance_Category']
num_cols = X_train.select_dtypes(include=['number']).columns.tolist()

# Ordinal order
education_order = ['Intermediate', 'Bachelors', 'Masters']
attendance_order = ['Low', 'Medium', 'High']

# Preprocessing
preprocessor = ColumnTransformer(transformers=[
    ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore'), onehot_cols),
    ('ordinal', OrdinalEncoder(categories=[education_order, attendance_order]), ordinal_cols),
    ('scaling', StandardScaler(), num_cols)
])

# Pipeline with Random Forest
pipeline = Pipeline(steps=[
    ('preprocessing', preprocessor),
    ('model', RandomForestRegressor(n_estimators=100, random_state=42))  # <-- Changed
])

# Train pipeline
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
import pandas as pd
comparison = pd.DataFrame({
    'Actual_Final_Score': y_test.values,
    'Predicted_Final_Score': y_pred
})
print(comparison.head(10))

from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("MAE:", mae)
print("RMSE:", rmse)
```
