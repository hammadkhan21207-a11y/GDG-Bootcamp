```python
df["Item"].fillna(df["Item"].mode()[0], inplace=True)
df["Quantity"].fillna(df["Quantity"].mean, inplace=True)
df["Price Per Unit"].fillna(df["Price Per Unit"].mode()[0], inplace=True)
df["Total Spent"].fillna(df["Total Spent"].mode()[0], inplace=True)
df["Payment Method"].fillna(df["Payment Method"].mode()[0], inplace=True)
df["Location"].fillna(df["Location"].mode()[0], inplace=True)
df["Transaction Date"].fillna(df["Transaction Date"].mode()[0], inplace=True)
```

MODIFIYING AND CLEANING

```python
for col in df.select_dtypes(include="object"):
    df[col] = df[col].astype(str).str.lower().str.strip()
```

```python
import numpy as np

df.replace(["N/A", "none", "-"], np.nan, inplace=True)
```

```python
df["Total Spent"] = pd.to_numeric(df["Total Spent"], errors="coerce")
median_value = df["Total Spent"].median()
df["Transaction Level"] = df["Total Spent"].apply(lambda x: "High" if x > median_value else "Low")
```

```python
df=df.drop_duplicates().reset_index(drop=True)
```

```python
df["Total Spent"] = pd.to_numeric(df["Total Spent"], errors="coerce")

item_summary = df.groupby("Item").agg(
    total_sales=pd.NamedAgg(column="Total Spent", aggfunc="sum"),
    average_transaction=pd.NamedAgg(column="Total Spent", aggfunc="mean")
).reset_index()

print(item_summary)
```
